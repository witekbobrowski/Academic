```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from scipy import stats
%matplotlib inline
sns.set()


def simulate_signals(time, frequency):
    queue = 0
    history = [queue]
    for _ in time:
        arrival = np.random.poisson(frequency)
        queue += arrival
        history.append(queue)
    return history
```

```python
time = range(0, 90)
lam = 1

[plt.plot(simulate_signals(time, lam)) for _ in range(0, 5)]
plt.show()
```

```python
def extract_signals(times, keys, time, frequency):
    results = {key:[] for key in keys}
    for _ in range(times):
        simulation = simulate_signals(time, frequency)
        for key in keys:
            results[key].append(simulation[key])
    return results

def poisson_dist(lam, t, k):
    return ((lam * t) ** k) / np.math.factorial(k) * np.exp(-lam * t)
```

```python
N = 10 ** 4
keys = [1, 20, 90]

results = extract_signals(N, keys, time, lam)

_ = [sns.distplot(results[key], kde=False) for key in results]
```
