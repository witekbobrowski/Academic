```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
%matplotlib inline


def simulate_signals(time, frequency):
    queue = 0
    history = [queue]
    for _ in time:
        arrival = np.random.poisson(frequency)
        queue += arrival
        history.append(queue)
    return history
```

```python
time = range(0, 90)
frequency = 1

[plt.plot(simulate_signals(time, frequency)) for _ in range(0, 5)]
plt.show()
```

```python
def extract_signals(times, keys, time, frequency):
    results = {key:[] for key in keys}
    for _ in range(times):
        simulation = simulate_signals(time, frequency)
        for key in keys:
            results[key].append(simulation[key])
    return results
```

```python
N = 10 ** 4
keys = [1, 20, 90]

results = extract_signals(N, keys, time, frequency)

[plt.plot(results[key]) for key in results]
plt.show()
```
