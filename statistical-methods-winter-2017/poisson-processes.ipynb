```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from scipy import stats
%matplotlib inline
sns.set()


def simulate_signals(max_time, lam):
    history = []
    t = 0
    while t < max_time:
        t += np.random.exponential(lam)
        history.append(t)
    return history
```

```python
max_time = 90
lam = 1

for _ in range(0, 5):
    results = simulate_signals(max_time, lam)
    plt.plot(results)
plt.show()
```

```python
def extract_signals(times, keys, time, frequency):
    results = {key:[] for key in keys}
    for _ in range(times):
        simulation = np.array(simulate_signals(time, frequency))
        for key in keys:
            results[key].append(len(simulation[simulation <= key]))
    return results

def poisson_dist(lam, t, k):
    return ((lam * t) ** k) / np.math.factorial(k) * np.exp(-lam * t)
```

```python
N = 10 ** 4
keys = [1, 20, 90]

results = extract_signals(N, keys, keys[-1], lam)

_ = [sns.distplot(results[key]) for key in keys]
```

```python
_ = [sns.distplot([poisson_dist(lam, key, k) for k in range(90)]) for key in keys]
```
